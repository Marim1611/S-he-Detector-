```python
import numpy as np
import math
import matplotlib.pyplot as plt
from skimage import io, color
```

```python
def gradient_info(img):
    height=img.shape[0]
    width=img.shape[1]
    magnitude=[]
    orientation=[]

    for j in range(height):
        mag_temp=[]
        dir_temp=[]
        for i in range(width):
            Gx=0
            Gy=0

            # compute Gx= Img(x+1,y)- Img(x-1,y)
            if i+1>=width and i>0:
                Gx= - img[j][i-1]
            elif i+1< width and i==0:
                Gx= img[j][i+1]
            else:
                Gx= img[j][i+1] - img[j][i-1]

            # compute Gy= Img(x,y+1)- Img(x,y-1)
            if j+1>=height and j>0:
                Gy= - img[j-1][i]
            elif j+1< height and j==0:
                Gy= img[j+1][i]
            else:
                Gy= img[j+1][i] - img[j-1][i]

            # compute magnitude and save it in the list 
            mag_temp.append(np.round(math.sqrt(Gx**2 + Gy**2),10))

            # compute angle and save it in the list
            if Gx==0: 
                dir_temp.append(np.round(math.degrees(0.0),10))
            elif Gy==0:
                dir_temp.append(np.round(math.degrees(-math.pi/2.0),10))
            else: 
                dir_temp.append(np.round(float(math.degrees(np.arctan(Gx/Gy))),10))
                    
        magnitude.append(mag_temp)
        orientation.append(dir_temp)  

    magnitude=np.array(magnitude)
    orientation=np.array(orientation)  

    return magnitude,orientation
```

```python
#-------------------- Testing gradient function --------------------------- 

img = np.array(color.rgb2gray(io.imread("F1/F1-10.jpg")))

gradient_magnitude,gradient_orientation=gradient_info(img)

plt.figure(1)
plt.imshow(img, cmap="gray")
plt.axis("off")
plt.title("Original image")

plt.figure(2)
plt.imshow(gradient_magnitude, cmap="gray")
plt.axis("off")
plt.title("Gradient magnitude")

plt.figure(3)
plt.imshow(gradient_orientation, cmap="gray")
plt.axis("off")
plt.title("Gradient orientation")
```

```python
# given an img divided into separated lines
# for each line, will divide it into 6 separate regions and compute gradient magnitude and orientation for each region
def divide_line_to_regions(img):

    width= img.shape[1]
    w1=int(np.floor(width/3)) 
    w2=width- w1
    w3=width

    height=img.shape[0]
    height_upper_region= int(np.floor(height/2)) 
    height_lower_region= height

    region1= img[0:height_upper_region,0:w1]
    region2= img[0:height_upper_region,w1:w2]
    region3= img[0:height_upper_region,w2:w3]
    region4= img[height_upper_region:height_lower_region,0:w1]
    region5= img[height_upper_region:height_lower_region,w1:w2]
    region6= img[height_upper_region:height_lower_region,w2:w3]
    
    return region1,region2,region3,region4,region5,region6
```

```python
#-------------------- Testing dividing image into 6 regions --------------------------- 

line_img = np.array(color.rgb2gray(io.imread("F1/F1-10.jpg")))
region1,region2,region3,region4,region5,region6= divide_line_to_regions(line_img)  

plt.figure(1,figsize=[10, 5])
plt.imshow(line_img, cmap="gray")
plt.title("Original image")
plt.axis("off")

fig, axs = plt.subplots(2, 3, figsize=[10, 3])
fig.suptitle('Image regions')

axs[0, 0].imshow(region1, cmap="gray")
axs[0, 0].axis('off')

axs[0, 1].imshow(region2, cmap="gray")
axs[0, 1].axis('off')

axs[0, 2].imshow(region3, cmap="gray")
axs[0, 2].axis('off')

axs[1, 0].imshow(region4, cmap="gray")
axs[1, 0].axis('off')

axs[1, 1].imshow(region5, cmap="gray")
axs[1, 1].axis('off')

axs[1, 2].imshow(region6, cmap="gray")
axs[1, 2].axis('off')
```

```python
def frequency_table(img_gradient_orientation):
    height= img_gradient_orientation.shape[0]
    width= img_gradient_orientation.shape[1]
    magnitude=np.zeros(9)

    for j in range(height):
        for i in range(width):
            if img_gradient_orientation[j][i] <20:
                magnitude[0]+=1
            elif img_gradient_orientation[j][i]<40 and img_gradient_orientation[j][i]>=20:
                magnitude[1]+=1
            elif img_gradient_orientation[j][i]<60 and img_gradient_orientation[j][i]>=40:
                magnitude[2]+=1
            elif img_gradient_orientation[j][i]<80 and img_gradient_orientation[j][i]>=60:
                magnitude[3]+=1
            elif img_gradient_orientation[j][i]<100 and img_gradient_orientation[j][i]>=80:
                magnitude[4]+=1
            elif img_gradient_orientation[j][i]<120 and img_gradient_orientation[j][i]>=100:
                magnitude[5]+=1
            elif img_gradient_orientation[j][i]<140 and img_gradient_orientation[j][i]>=120:
                magnitude[6]+=1
            elif img_gradient_orientation[j][i]<160 and img_gradient_orientation[j][i]>=140:
                magnitude[7]+=1
            else:
                magnitude[8]+=1
           
    return magnitude
```

```python
# ---------------------- Testig histogram function ---------------------------


magnitude1=frequency_table(region1_gradient_orientation)
magnitude2=frequency_table(region2_gradient_orientation)
magnitude3=frequency_table(region3_gradient_orientation)
magnitude4=frequency_table(region4_gradient_orientation)
magnitude5=frequency_table(region5_gradient_orientation)
magnitude6=frequency_table(region6_gradient_orientation)

bins=list(range(0, 180,20))

# region 1 histogram
fig = plt.figure(figsize = (10, 5))
plt.bar(bins, magnitude1)
plt.title("Region 1")

# region 2 histogram
fig = plt.figure(figsize = (10, 5))
plt.bar(bins, magnitude2)
plt.title("Region 2")


# region 3 histogram
fig = plt.figure(figsize = (10, 5))
plt.bar(bins, magnitude3)
plt.title("Region 3")


# region 4 histogram
fig = plt.figure(figsize = (10, 5))
plt.bar(bins, magnitude4)
plt.title("Region 4")


# region 5 histogram
fig = plt.figure(figsize = (10, 5))
plt.bar(bins, magnitude5)
plt.title("Region 5")


# region 6 histogram
fig = plt.figure(figsize = (10, 5))
plt.bar(bins, magnitude6)
plt.title("Region 6")

# concatenated histogram 
# magnitude=np.zeros(9)
# magnitude[0]=magnitude1[0]+magnitude2[0] 


```
